```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from flask import Flask, Response, render_template
import threading
import time

from PIL import Image
import requests  
```

### 1.Reading Image

```python
img = cv2.imread('image.jpg')
```

### 2.Showing Image

```python
cv2.imshow('image',img) #OpenCV creates a window but it cannot update, refresh, or respond
cv2.waitKey() # # is required because OpenCV does not automatically keep the window alive.
```

### 3.To Display Image to Console

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
print('Height of image:', img.shape[0], 'pixels')
print('Width of image:', img.shape[1], 'pixels')
```

### 4. Write (save) image

```python
cv2.imwrite("output_image.jpg", img)
print("Image saved as output_image.jpg")
```

### 5. Close windows

```python
cv2.destroyAllWindows() # After you close the window, this cleans up the GUI resources.
```
